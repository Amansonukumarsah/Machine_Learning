# DATA VISUALIZATION

histogram is representation of data distribution

```python
df_vis = df_copy.hist(figsize=(20,20),bins=20)
for ax in df_vis.flatten():
    ax.set_title(ax.get_title(),fontsize=16)
    ax.set_xlabel("Frequency",fontsize=14)
    ax.set_ylabel("Number_Of_People",fontsize=14)
plt.tight_layout()
plt.show()
```

```python
df_copy['Pregnancies'].fillna(df_copy['Pregnancies'].mean(),inplace = True)
df_copy['Glucose'].fillna(df_copy['Glucose'].mean(),inplace=True)
df_copy['BloodPressure'].fillna(df_copy['BloodPressure'].median(),inplace=True)
df_copy['SkinThickness'].fillna(df_copy['SkinThickness'].median(),inplace=True)
df_copy['Insulin'].fillna(df_copy['Insulin'].median(),inplace=True)
df_copy['BMI'].fillna(df_copy['BMI'].median(),inplace=True)
```

```python
df_vis = df_copy.hist(figsize=(20,20),bins=20)
for ax in df_vis.flatten():
    ax.set_title(ax.get_title(),fontsize=16)
    ax.set_xlabel("Frequency",fontsize=14)
    ax.set_ylabel("Number_Of_People",fontsize=14)
plt.tight_layout()
plt.show()
```
