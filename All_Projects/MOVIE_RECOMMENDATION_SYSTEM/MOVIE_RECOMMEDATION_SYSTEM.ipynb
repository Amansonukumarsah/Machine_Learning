```python
#'[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'
#['Action','Adventure','Fantasy','Science Fiction']

def convert(obj):
    count=0
    arr=[]
    for i in ast.literal_eval(obj):
        arr.append(i["name"])
    return arr
def convert1(obj):
    arr=[]
    count=0
    for i in ast.literal_eval(obj):
        if count<=2:
            arr.append(i["name"])
            count+=1
        else:
            break
    return arr
def convert2(obj):
    arr=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            arr.append(i["name"])
            break
    return arr
    
```

```python
movies_data['genres'] = movies_data['genres'].apply(convert)
movies_data['keywords'] = movies_data['keywords'].apply(convert)
movies_data['crew'] = movies_data['crew'].apply(convert2)
movies_data['cast'] = movies_data['cast'].apply(convert1)
```

```python
#NOW APPLYING THE VICTORIZATION :- IT IS USED TO CONVERT THE TEXT INOT VECTOR AND THEN HE FIND THE CLOSET VECTOR USING GRPAH
```

```python
import nltk
```

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    arr=[]
    for i in text.split():
        arr.append(ps.stem(i))
    return " ".join(arr)
        
```

```python
new_movie_data['tags'] = new_movie_data['tags'].apply(stem)
df = new_movie_data[['movie_id','title','tags']]
```

```python
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(df['tags']).toarray()
```

```python
cv.get_feature_name()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity.shape
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])
```

```python
enumerate(similarity)
```

```python
def Movie_Recommended(movie):
    movie_index = df[df['title']==movie].index[0]
    distance = similarity[movie_index]
    movie_list = sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1])[1:6]
    for i in movie_list:
        print(df.iloc[i[0]].title)
    return
```
