```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('titanic.csv')
```

```python
df.head()
```

```python
df['number'].unique()
```

```python
fig = df['number'].value_counts().plot.bar()
fig.set_title('Passengrs Travelling Data')
```

```python
# extract numeric data
df['numeric_data'] = pd.to_numeric(df['number'], errors='coerce',downcast='integer')
```

```python
df
```

```python
# extract categorical data
df['categorical_data'] = np.where(df['numeric_data'].isnull(),df['number'],np.nan)
```

```python
df
```

```python
df['Cabin'].unique()
```

```python
df['cabin_number'] = df['Cabin'].str.extract('(\d+)')
# Extract the numerical data
df['cabin_categorical'] = df['Cabin'].str[0]
```

```python
df.head()
```

```python
df['numric_ticket_number'] = df['Ticket'].apply(lambda s:s.split()[-1])
```

```python
df.info()
```

```python
df['numric_ticket_number'] = pd.to_numeric(df['numric_ticket_number'], errors='coerce',downcast='integer')
```

```python
df
```

```python
df.info()
```

```python
df['ticket_cat'] = df['Ticket'].apply(lambda s:s.split()[0])
```

```python
df['ticket_cat'] = np.where(df['ticket_cat'].str.isdigit(),np.nan,df['ticket_cat'])
```
